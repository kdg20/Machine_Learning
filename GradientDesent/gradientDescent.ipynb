```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
```

```python
#m is slope and b is intercept
#gradient descent finds best fit line
#m is slope and b is intercept which is initially 0 and we will increse this to achieve global minima


def gradient_Descent(x,y):
    m_current=b_current=0;
    iterations=1000;
    n=len(x);
    learning_rate=0.008;
    
    for i in range(iterations):
        y_predicted=m_current*x + b_current
        cost=(1/n)*sum([val**2 for val in (y-y_predicted)])
        nd= -2/n * sum(x*(y-y_predicted))
        bd= -2/n * sum((y-y_predicted))
        
        m_current=m_current-learning_rate*nd
        b_current=b_current-learning_rate*bd
        
        print("m {},b {}, cost {}, iterations".format(m_current,b_current,cost,i))
    pass

gradient_Descent(x,y)
```
